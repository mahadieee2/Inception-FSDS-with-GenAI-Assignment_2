# 1. Email Validator (String Filtering + Loops)

```python
""" 
Task: Check if a given email is valid — must contain “@” and end with “.com” or “.org”.
Example:
Input: hello@inceptionbd.com
Output: Valid Email

"""

email = input("Enter your email: ")

# Check if email contains '@' and ends with '.com' or '.org'
if "@" in email and (email.endswith(".com") or email.endswith(".org")):
    print("Valid Email")
else:
    print("Invalid Email")
```

# 2. Count Word Frequency in Customer Reviews

```python
""" 
Task: Given a review text, count how many times each word appears (ignore case).
Example:
Input: "This product is good and this service is excellent"
Output: {'this': 2, 'product': 1, 'is': 2, 'good': 1, 'and': 1, 'service': 1, 'excellent': 1}

"""

text = "This product is good and this service is excellent"

text = text.lower()
A = text.split()

C = {}
for i in A:
    if i in C:
        C[i] += 1
    else:
        C[i] = 1

print(C)
```

# 3. Password Strength Checker

```python
"""  
Task: Check if a password is strong — must have at least one uppercase, one lowercase, one digit, and one special character.
Example:
Input: "Hello@123"
Output: Strong Password


"""

password = input("Enter your password here: ")

if (any(i.islower() for i in password) and
    any(i.isupper() for i in password) and
    any(i.isdigit() for i in password) and
    any(i in "!@#$%^&*()-_=+[]{}|;:',.<>?/" for i in password)):
    print("Strong Password")
else:
    print("Weak Password")
```

# 5. SMS Word Counter

```python

""" 
Task: Count how many characters and words are in a message, and tell if it exceeds 160 characters (SMS limit).
Example:
Input: "This is a demo message."
Output: Words = 5, Characters = 23, Status = Within Limit


"""



message = input("Enter your SMS message: ")

char_count = len(message)

words = message.split()
word_count = len(words)

# Check limit
if char_count > 160:
    status = "Exceeded Limit"
else:
    status = "Within Limit"

print("Words =", word_count)
print("Characters =", char_count)
print("Status =", status)
```

# 6. Remove Duplicates from Contact List

```python
""" 
Task: Given a list of names (comma separated), remove duplicates.
Example:
Input: "Rahim, Karim, Rahim, Sultana"
Output: "Rahim, Karim, Sultana"

"""

contacts = input("Enter names separated by commas: ")

items = contacts.split(",")

unique_list = []
for name in items:
    clean_name = name.strip() 
    if clean_name not in unique_list:
        unique_list.append(clean_name)


result = ", ".join(unique_list)

print(result)
```

# 7. Detect Suspicious Words in Text (Moderation)

```python
""" 
Task: If any “banned” words like ['spam', 'scam', 'fake'] are found in a text, print “Warning”.
Example:
Input: "This is a scam offer"
Output: Warning: Suspicious content detected


"""


text = input("Enter your text: ")

banned_words = ['spam', 'scam', 'fake']
warning = False

# Convert to lowercase for accurate detection
lower_text = text.lower()

for word in banned_words:
    if word in lower_text:
        warning = True
        break

if warning:
    print("Warning: Suspicious content detected")
else:
    print("Text is clean")
```

# 8. Invoice Code Formatter

```python
""" 
Task: Given an invoice number like 23, generate code in the format INV00023.
Example:
Input: 23
Output: INV00023

"""

num = input("Enter invoice number: ")

# Add leading zeros using rjust
formatted = "INV" + num.rjust(5, "0")

print(formatted)
```

# 9. Count Hashtags and Mentions (Social Media Analytics)

```python
""" 
Task: Count how many hashtags (#) and mentions (@) appear in a post.
Example:
Input: "Check out #AIWorkshop and follow @inceptionbd"
Output: Hashtags = 1, Mentions = 1


"""




post = input("Enter your social media post: ")

hashtag_count = 0
mention_count = 0

for i in post:
    if i == "#":
        hashtag_count += 1
    elif i == "@":
        mention_count += 1

print("Hashtags =", hashtag_count)
print("Mentions =", mention_count)
```

# 10. Word Reversal for Data Encryption

```python
""" 
Task: Reverse each word in a sentence while keeping word order the same.
Example:
Input: "AI is fun"
Output: "IA si nuf"

"""

sentence = input("Enter a sentence: ")

words = sentence.split()
reversed_words = []

for w in words:
    reversed_word = ""
    for ch in w:
        reversed_word = ch + reversed_word   # reverse manually
    reversed_words.append(reversed_word)

result = " ".join(reversed_words)
print(result)
```
